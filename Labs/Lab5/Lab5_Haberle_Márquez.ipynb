Sin etiquetas reales no pueden aplicarse m√©tricas tradicionales, debiendo recurrir a indicadores alternativos que suelen ser subjetivos y no siempre fiables. Adem√°s, los grupos generados por el algoritmo no necesariamente corresponden a categor√≠as √∫tiles o representativas de la realidad, lo que dificulta validar su relevancia.
3. La matriz de confusi√≥n es una tabla que permite evaluar el desempe√±o de un modelo de clasificaci√≥n comparando las etiquetas reales con las predicciones del modelo. En ella, las filas representan las clases reales y las columnas las clases predichas (o viceversa), mostrando cu√°ntas observaciones fueron correctamente clasificadas y cu√°ntas se confundieron entre categor√≠as.
A partir de esta matriz se pueden calcular m√©tricas clave como precisi√≥n (precision), recall o sensibilidad, especificidad, exactitud (accuracy), entre otras.
4. * Accuracy se calcula como (TP + TN) / (TP + TN + FP + FN) y mide la proporci√≥n de aciertos totales del modelo, siendo √∫til cuando las clases est√°n balanceadas, pero puede ser enga√±osa si una clase es mucho m√°s frecuente que la otra.
    * Precision se calcula como TP / (TP + FP) y representa qu√© tan confiables son las predicciones positivas, es decir, de todos los casos que el modelo clasific√≥ como positivos, cu√°ntos realmente lo eran.

    * Recall se calcula como TP / (TP + FN) y mide la capacidad del modelo para encontrar todos los casos positivos reales, siendo clave cuando es m√°s importante no dejar pasar positivos sin detectar (Por ejemplo el ejemplo cl√°sico de detecci√≥n de enfermedades).

    * F1 score se calcula como 2 * (Precision * Recall) / (Precision + Recall). As√≠, vemos que es una ponderaci√≥n entre precisi√≥n y recall, siendo especialmente √∫til cuando se busca un balance entre evitar falsos positivos y no perder falsos negativos.
5. M√©tricas recomendadas:
    1. M√©trica recomendada: Recall. En este escenario es cr√≠tico no dejar pasar fallas reales (falsos negativos), ya que una falla no detectada puede generar altos costos y riesgos de seguridad.
    2. M√©trica recomendada: Recall. Aqu√≠ lo m√°s importante es identificar a todos los enfermos, incluso si eso implica tener algunos falsos positivos. No detectar a una persona contagiada puede propagar la enfermedad.
    3. M√©trica recomendada: Precision. En este caso, aprobar err√≥neamente un cr√©dito a alguien que no lo pagar√° (falso positivo) tiene un alto costo econ√≥mico.
    4. M√©trica recomendada: F1 score. La detecci√≥n de cr√≠menes requiere un equilibrio: no dejar pasar delitos (recall) pero tambi√©n evitar falsas acusaciones (precision).
6. La calibraci√≥n consiste en ajustar las probabilidades que entrega un modelo sobre la clasificaci√≥n de una cierta etiqueta con respecto a la probabilidad real de que esa etiqueta realmente ocurra. Tienen principal importancia en contextos donde los eventos o categor√≠as predichas son poco frecuentes o son muy importantes de ser acertados, y se usa pues si bien un modelo puede predecir bien, no siempre se condice con la realidad.

2. Hecho esto, ¬øcu√°ntos se tienen ejemplos por cada clase? Comente lo que observa. [1 punto]

3. Genere un `ColumnTransformer` en donde especifique las transformaciones que hay que realizar para cada columna (por ejemplo StandarScaler, MinMaxScaler, OneHotEncoder, etc...) para que puedan ser utilizadas correctamente por el modelo predictivo y gu√°rdelo una variable llamada `col_transformer`. **Respuesta:**

### 1

````python
import pandas as pd
df = pd.read_csv(path+"/stats_players.csv")
df_label = df.copy()
df_label["label"] = df["National_Position"].apply(lambda x: 0 if  pd.isna(x) else 1)
df_label.head()
````

### 2

````python
df_label["label"].value_counts()
````

Vemos que hay 16513 valores nulos, representando as√≠ la clase 0, de jugaodres no seleccionados para jugar. Por su parte, los jugadores seleccionados tienen 1075 ejemplos con label = 1, siendo l√≥gico, ya que son pocos los jugadores que representan a su pa√≠s vs el total de jugadores.

### 3

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler

df_transform = df_label.copy().drop(columns=["National_Position", "Name"])
ordinal = ["Weak_foot","Skill_Moves"]
numeric_features = df_transform.select_dtypes(include="number").drop(columns=ordinal).drop(columns=["label"]).columns.tolist()
categorical = df_transform.select_dtypes("object").columns.tolist()

col_transformer = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("categorical", OneHotEncoder(handle_unknown="ignore"), categorical),
        ("ordinal", MinMaxScaler(), ordinal)
    ]
)
````

### 4

De partida, eliminamos nation_position y name, ya que la primera implicar√≠a un data_leakage sobre la variable objetivo label, y name no aporta nada al modelo aparte de tenre una cardinalidad enorme.

Viendo el html provisto, la gran mayor√≠a de los atributos son valores num√©ricos entre 0 y 100 aproximadamente (Peso y altura difieren pero est√°n en la misma escala aprox), adem√°s de distribuir casi como una normal. De esta forma, simplemente aplicamos un estandar scaler sobre estos. Sin embargo, hay un par de valores num√©ricos que son ordinales, siendo estos Weak_foot y Skill_moves, representando la habilidad de cada jugador sobre cada una, con un orden definido (0 muy malo, 5 muy bueno). Para esto, apicamos MinMaxScaler. Por √∫ltimo, a los categoricals, aplicamos un one hot encoder para cada una, teniendo as√≠ columnas binarias para cada valor.

### 1.2 Entrenamiento [3 puntos]

Ahora, vamos a entrenar los pipelines generados en los pasos anteriores. Para esto, debe realizar las siguientes tareas:

1. Separe los datos de entrenamiento en un conjunto de entrenamiento y de prueba  (la proporci√≥n queda a su juicio). En este paso, seleccione los ejemplos de forma aleatoria e intente mantener la distribuci√≥n original de labels de cada clase en los conjuntos de prueba/entrenamiento. (vea la documentaci√≥n de `train_test_split`). [1 puntos]


2. Defina un pipeline llamado `pipeline_xgboost` y otro llamado `pipeline_lightgbm`. Estos pipelines deben tener el mismo ColumnTransformer definido en la secci√≥n de preprocesamiento, pero deben variar los clasificadores de acuerdo al nombre de cada pipeline. [1 puntos]

3. Entrene los pipelines. [1 punto]



**Respuesta:**

````python
from sklearn.model_selection import train_test_split

X = df_transform.drop(columns=["label"])
y = df_transform["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1892, stratify=y
)
````

Analizando la distribuci√≥n de la variable objetivo en el punto anterior, tenemos aproximadamente 3500 datos aprox para validar nuestros resultados y 14 mil datos de entrenamiento. Viendo que hay casi mil ejemplos de la variable objetivo, con casi el triple de datos de validacion con respecto a ese numero, debiesemos estar bien en especial al mantener las proporciones con stratify.

````python
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

pipeline_xgboost = Pipeline(steps=[
    ("preprocess", col_transformer),
    ("clf", XGBClassifier(
        random_state=1892,
    ))
])

pipeline_lightgbm = Pipeline(steps=[
    ("preprocess", col_transformer),
    ("clf", LGBMClassifier(
        random_state=1892
    ))
])
````

````python
# Entrenar XGBoost
pipeline_xgboost.fit(X_train, y_train)
````

````python
# Entrenar LightGBM
pipeline_lightgbm.fit(X_train, y_train)
````

**Respuesta:**

### 1.

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, f1_score

# Predicciones
y_pred_xgb = pipeline_xgboost.predict(X_test)
y_pred_lgb = pipeline_lightgbm.predict(X_test)

# XGBoost
acc_xgb = accuracy_score(y_test, y_pred_xgb)
prec_xgb = precision_score(y_test, y_pred_xgb)
rec_xgb = recall_score(y_test, y_pred_xgb)

print("XGBoost")
print("Accuracy:", acc_xgb)
print("Precision:", prec_xgb)
print("Recall:", rec_xgb)
print(classification_report(y_test, y_pred_xgb))

# LightGBM
acc_lgb = accuracy_score(y_test, y_pred_lgb)
prec_lgb = precision_score(y_test, y_pred_lgb)
rec_lgb = recall_score(y_test, y_pred_lgb)

print("\nLightGBM")
print("Accuracy:", acc_lgb)
print("Precision:", prec_lgb)
print("Recall:", rec_lgb)
print(classification_report(y_test, y_pred_lgb))
````

Influencia del desbalance: la gran diferencia en la cantidad de ejemplos (clase 0 ‚â´ clase 1) hace que el modelo aprenda a priorizar la clase mayoritaria para maximizar accuracy, sacrificando la detecci√≥n de los pocos casos positivos, lo que explica el recall tan bajo.

### 3.

La elecci√≥n del clasificador se justificar en base a la siguiente [gu√≠a](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) y se deben comentar los resultados obtenidos en la clasificaci√≥n.

**Tareas:** [1 punto por tarea]

**Respuesta:**

### 1.

````python
# Agregamos algunas posiciones faltantes seg√∫n nuestra l√≥gica y lo de la gu√≠a

ataque = ['ST', 'CF', 'LF', 'RF', 'LS', 'RS']
central_ataque = ['RW', 'CAM', 'LW', 'RAM', 'LAM']
central = ['RM', 'CM', 'LM', 'RCM', 'LCM']
central_defensa = ['RWB', 'CDM', 'LWB', 'LDM', 'RDM']
defensa = ['RB', 'CB', 'LB', 'LCB', 'RCB']
arquero = ['GK']
````

````python
def assign_group(pos):
  if pos in ataque:
    return "ataque"
  elif pos in central_ataque:
    return "central_ataque"
  elif pos in central:
    return "central"
  elif pos in central_defensa:
    return "central_defensa"
  elif pos in defensa:
    return "defensa"
  elif pos in arquero:
    return "arquero"

df_pos = df_label.copy().dropna(subset=["Club_Position"])
df_pos = df_pos[df_pos["Club_Position"] != "Sub"]
df_pos = df_pos[df_pos["Club_Position"] != "Res"]
df_pos["label"] = df_pos["Club_Position"].apply(assign_group)
df_pos.head()
````

````python
df_pos["Club_Position"].unique()
````

### 2.

````python
df_pos["label"].value_counts()
````

### 3.

````python
df_pos_transform = df_pos.copy().drop(columns=["National_Position", "Name", "Club_Position"])
ordinal = ["Weak_foot","Skill_Moves"]
numeric_features = df_pos_transform.select_dtypes(include="number").drop(columns=ordinal).columns.tolist()
categorical = df_pos_transform.select_dtypes("object").drop(columns=["label"]).columns.tolist()

pos_col_transformer = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("categorical", OneHotEncoder(handle_unknown="ignore"), categorical),
        ("ordinal", MinMaxScaler(), ordinal)
    ]
)
````

````python
X = df_pos_transform.drop(columns=["label"])
y = df_pos_transform["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1892, stratify=y
)
````

````python
from sklearn.svm import LinearSVC

pipeline_linear_svc = Pipeline(steps=[
    ("preprocess", pos_col_transformer),
    ("svc", LinearSVC())
])
````

````python
pipeline_linear_svc.fit(X_train, y_train)
````

### 4.

````python
# Predicciones
y_pred_svc = pipeline_linear_svc.predict(X_test)

# XGBoost
prec_svc = precision_score(y_test, y_pred_svc, average="weighted")
rec_svc = recall_score(y_test, y_pred_svc, average="weighted")
f1_svc = f1_score(y_test, y_pred_svc, average="weighted")

print("Linear_SVC")
#print("Accuracy:", acc_svc)
print("Precision:", prec_svc)
print("Recall:", rec_svc)
print("F1 Score:", f1_svc)
print(classification_report(y_test, y_pred_svc))
````

En primera instancia, elegimos un modelo LinearSVC seg√∫n el recurso de elecci√≥n de modelos provisto por el enunciado y las caracter√≠sticas del dataset. Con esto, vemos que los resultados demuestran un mucho mejor desempe√±o con respecto al modelo anterior. Vemos que clasifica muy bien los arqueros, siendo algo l√≥gico pues si bien son muy pocos tienen una distribuci√≥n de atributos completamente distinta a los otros jugadores. Tambi√©n se desempe√±a bien en defensa y ataque, que son clases con muchos ejemplos y relativamente homog√©neas, por lo que el modelo tiene suficiente informaci√≥n para aprenderlas correctamente.

En cambio, las posiciones centrales y las combinadas como central_defensa y central_ataque son m√°s dif√≠ciles. Esto refleja dos factores: primero, hay menos ejemplos de estas clases en el dataset, lo que dificulta que el modelo aprenda patrones claros; segundo, futbol√≠sticamente son roles h√≠bridos que combinan caracter√≠sticas de defensa y ataque, por lo que sus estad√≠sticas se parecen entre s√≠ y generan confusi√≥n para el modelo.

En resumen, el desempe√±o del modelo est√° muy ligado al desbalance de clases: las posiciones con m√°s datos y m√°s definidas se predicen bien, mientras que las clases con menos ejemplos o con roles intermedios tienen baja precisi√≥n y recall, lo que limita la capacidad de diferenciar correctamente a todos los jugadores.

### 2.
* dt (Decision Tree ‚Äì √Årbol de Decisi√≥n)
Es un modelo que divide los datos en ramas jer√°rquicas de decisiones basadas en las caracter√≠sticas (features). En cada nodo se elige la variable y el umbral que mejor separa las clases (por ejemplo, usando Gini o entropy). El resultado es un √°rbol en el que cada hoja representa una predicci√≥n.

* rf (Random Forest ‚Äì Bosque Aleatorio)
Es un ensamble de muchos √°rboles de decisi√≥n independientes. Cada √°rbol se entrena con una muestra aleatoria de los datos y un subconjunto de caracter√≠sticas, y la predicci√≥n final se obtiene por votaci√≥n mayoritaria (clasificaci√≥n) o promedio (regresi√≥n). Esto reduce el sobreajuste t√≠pico de un solo √°rbol y mejora la precisi√≥n.

* et (Extra Trees ‚Äì Extremely Randomized Trees)
Similar a Random Forest, pero introduce m√°s aleatoriedad al elegir los puntos de corte en cada nodo, no solo las variables. Al hacer divisiones m√°s aleatorias, genera √°rboles m√°s diversos y puede mejorar la generalizaci√≥n y la velocidad de entrenamiento.

### 1.

````python
X = df_transform.drop(columns=["label"])
y = df_transform["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1892, stratify=y
)
````

````python
pipeline_xgb_full = Pipeline(steps=[
    ("preprocess", col_transformer),
    ("clf", XGBClassifier(random_state=1892))
])

pipeline_xgb_full.fit(X_train, y_train)
y_pred_full = pipeline_xgb_full.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_full))
print(classification_report(y_test, y_pred_full))
````

### 2.

````python
import plotly.express as px

def plot_feature_importance(pipeline_xgb_full, importance_type = "weight"):
  feature_names = pipeline_xgb_full.named_steps['preprocess'].get_feature_names_out()
  importance_dict = pipeline_xgb_full.named_steps['clf'].get_booster().get_score(importance_type=importance_type)

  importance_named = {}
  for k, v in importance_dict.items():
      fname = feature_names[int(k[1:])]
      cleaned_name = "__".join(fname.split("__")[1:]) if "__" in fname else fname
      importance_named[cleaned_name] = v

  importance_df = pd.DataFrame({
      'Feature': list(importance_named.keys()),
      'Importance': list(importance_named.values())
  }).sort_values(by='Importance', ascending=False)

  fig = px.bar(
      importance_df,
      x="Importance",
      y="Feature",
      orientation="h",
      title=f"XGBoost Feature Importance (by {importance_type})",
      height=800
  )
  fig.update_layout(yaxis=dict(autorange="reversed"),
                    margin=dict(l=250),)
  fig.show()
````

````python
plot_feature_importance(pipeline_xgb_full, importance_type="weight")
````

### 3.

````python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

feature_names = pipeline_xgb_full.named_steps['preprocess'].get_feature_names_out()

X_train_trans = pipeline_xgb_full.named_steps['preprocess'].transform(X_train)
X_test_trans = pipeline_xgb_full.named_steps['preprocess'].transform(X_test)

if not isinstance(X_train_trans, np.ndarray):
    X_train_trans = X_train_trans.toarray()
    X_test_trans = X_test_trans.toarray()

X_train_df = pd.DataFrame(X_train_trans, columns=feature_names)
X_test_df = pd.DataFrame(X_test_trans, columns=feature_names)

base_model = pipeline_xgb_full.named_steps['clf']

base_model = pipeline_xgb_full.named_steps['clf']
base_model.fit(X_train_df, y_train)

importance_dict = base_model.get_booster().get_score(importance_type="weight")

feat_imp = pd.DataFrame({
    "Feature": list(importance_dict.keys()),
    "Importance": list(importance_dict.values())
}).sort_values("Importance", ascending=False).reset_index(drop=True)

results = []
for k in range(10, len(feat_imp)+1, 10):
    selected = feat_imp.head(k)["Feature"].tolist()

    model = XGBClassifier(random_state=1892)
    model.fit(X_train_df[selected], y_train)
    y_pred = model.predict(X_test_df[selected])

    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results.append((k, prec, rec, f1))
    print(f"Top {k} features ‚Üí Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

results_df = pd.DataFrame(results, columns=["Num_Features", "Precision", "Recall", "F1"])
````

### 4.

````python
from sklearn.metrics import classification_report, confusion_matrix

N = 80
selected_features = feat_imp.head(N)["Feature"].tolist()

final_model = XGBClassifier(random_state=1892)
final_model.fit(X_train_df[selected_features], y_train)

y_pred_reduced = final_model.predict(X_test_df[selected_features])

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_reduced))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_reduced, target_names=["No Seleccionado", "Seleccionado"]))
````

Para lograr modelos m√°s modulares, se recomienda realizar una calibraci√≥n del modelo entrenado anteriormente, con el objetivo de obtener salidas que reflejen mayor modularidad.

1. Se solicita que utilice un m√©todo de calibraci√≥n que asegure que las probabilidades generadas incrementen de manera mon√≥tona. Una m√©trica ampliamente utilizada para evaluar la precisi√≥n de la calibraci√≥n de un modelo es el Brier Score. Calcule el Brier Score para el modelo tanto antes como despu√©s de la calibraci√≥n. Esto le permitir√° realizar una comparaci√≥n cuantitativa y determinar si la calibraci√≥n ha mejorado el rendimiento del modelo. Para m√°s informaci√≥n sobre el Brier Score, puede consultar el siguiente enlace: [Scikit-Learn - Brier Score Loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html). [3 puntos]

2. Tras la calibraci√≥n, examine y comente los resultados obtenidos. A su an√°lisis a√±ada una comparaci√≥n visual de las ideales versus las salidas del modelo original (sin calibrar) y del modelo calibrado. [3 puntos]

**Respuesta**

````python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import brier_score_loss

# Probabilidades antes de calibraci√≥n
y_proba_uncal = final_model.predict_proba(X_test_df[selected_features])[:, 1]
brier_uncal = brier_score_loss(y_test, y_proba_uncal)
print("Brier Score (sin calibrar):", brier_uncal)

# Calibraci√≥n con isotonic regression
calibrated_model = CalibratedClassifierCV(
    estimator=final_model, method='isotonic'
)
calibrated_model.fit(X_train_df[selected_features], y_train)

# Probabilidades calibradas
y_proba_cal = calibrated_model.predict_proba(X_test_df[selected_features])[:, 1]
brier_cal = brier_score_loss(y_test, y_proba_cal)
print("Brier Score (calibrado):", brier_cal)
````

````python
from sklearn.calibration import calibration_curve
import plotly.graph_objects as go

# Compute calibration curves
prob_true_uncal, prob_pred_uncal = calibration_curve(y_test, y_proba_uncal, n_bins=20)
prob_true_cal, prob_pred_cal = calibration_curve(y_test, y_proba_cal, n_bins=20)

# Build Plotly figure
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=prob_pred_uncal, y=prob_true_uncal,
    mode="lines+markers",
    name="Sin calibrar"
))

fig.add_trace(go.Scatter(
    x=prob_pred_cal, y=prob_true_cal,
    mode="lines+markers",
    name="Calibrado (Isotonic)"
))

# Perfect calibration line
fig.add_trace(go.Scatter(
    x=[0, 1], y=[0, 1],
    mode="lines",
    line=dict(dash="dash", color="black"),
    name="Perfectamente calibrado"
))

fig.update_layout(
    title="Curvas de calibraci√≥n",
    xaxis_title="Probabilidad predicha",
    yaxis_title="Probabilidad real",
    width=700,
    height=600
)

fig.show()
````

#### Resultados tras calibraci√≥n

Brier Score sin calibrar: 0.0367

Brier Score calibrado: 0.0356

La reducci√≥n, aunque peque√±a, indica que el modelo calibrado produce probabilidades m√°s consistentes con la realidad (menor error cuadr√°tico medio entre probabilidad predicha y etiqueta real).

#### Interpretaci√≥n Visual

* Modelo sin calibrar: Tiende a estar razonablemente cerca de la diagonal en rangos medios de probabilidad. Sin embargo, en los extremos altos (prob > 0.8) el modelo es demasiado confiado, predice probabilidades muy altas para casos que en realidad no se cumplen con la misma frecuencia. Esto se refleja en un sesgo hacia la sobreconfianza.

* Modelo calibrado con Isotonic Regression: Ajusta las probabilidades para alinearlas mejor con la frecuencia observada. En general, se acerca m√°s a la diagonal ideal, especialmente cuando las probabilidades son relativamente bajas (< 0.7). En los valores extremos puede verse cierta sobrecorrecci√≥n (ej. ca√≠da en 0.9‚Äì1.0), lo que es esperable si hay pocos datos en esos bins.

* Curva ideal:Representa un modelo perfectamente calibrado: cada probabilidad predicha corresponde exactamente a la frecuencia observada. Ni el modelo original ni el calibrado alcanzan esta curva, pero el calibrado se aproxima m√°s en varios tramos.

#### Conclusi√≥n

La calibraci√≥n mejora levemente el desempe√±o del modelo seg√∫n el Brier Score Asimismo, la curva de calibraci√≥n muestra que el modelo calibrado entrega probabilidades m√°s fiables, cuando la probabilidad real no es muy alta. Por √∫ltimo, un modelo calibrado permite interpretar las probabilidades como riesgos reales, facilitando decisiones de negocio m√°s seguras.
