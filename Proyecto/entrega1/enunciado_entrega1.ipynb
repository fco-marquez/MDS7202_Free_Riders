#### Craci√≥n variable week

```python
# Transformar id's a string's
df_customers["customer_id"] = df_customers["customer_id"].astype(str)
df_products["product_id"] = df_products["product_id"].astype(str)
df_transactions["customer_id"] = df_transactions["customer_id"].astype(str)
df_transactions["product_id"] = df_transactions["product_id"].astype(str)
df_transactions["order_id"] = df_transactions["order_id"].astype(str)

# Crear variable week
df_transactions["week"] = df_transactions["purchase_date"].dt.isocalendar().week

df_grouped = df_transactions.copy()
# Crear variable objetivo
df_grouped['bought'] = 1
df_grouped['bought'] = df_grouped['bought'].astype('int8')
print(df_grouped.info())
df_grouped.head()
```

#### Creaci√≥n variables objetivos negativas (0)

```python
# Transformaci√≥n de tipo para ahorrar memoria

def optimize_numeric_types(df: pd.DataFrame) -> pd.DataFrame:
    """
    Convierte de una sola vez:
    - int64  -> int32
    - float64 -> float32
    Mantiene el resto igual.
    """
    dtypes = df.dtypes
    mapping = {}

    # Selecci√≥n vectorizada con numpy
    mask_int64   = dtypes.values == np.dtype("int64")
    mask_float64 = dtypes.values == np.dtype("float64")

    if mask_int64.any():
        mapping.update(dict(zip(dtypes.index[mask_int64], np.repeat("int32", mask_int64.sum()))))
    if mask_float64.any():
        mapping.update(dict(zip(dtypes.index[mask_float64], np.repeat("float32", mask_float64.sum()))))

    return df.astype(mapping, copy=False)


# IDs como category
for frame, cols in [
    (df_customers, ["customer_id", "customer_type"]),
    (df_products,  ["product_id", "brand", "category", "sub_category", "segment", "package"]),
    (df_transactions, ["customer_id", "product_id"]),
    (df_grouped,   ["customer_id", "product_id"]),
]:
    frame[cols] = frame[cols].astype("category")

# Optimizaci√≥n num√©rica 
df_transactions = optimize_numeric_types(df_transactions)
df_grouped       = optimize_numeric_types(df_grouped)
df_customers    = optimize_numeric_types(df_customers)
df_products     = optimize_numeric_types(df_products)
```

```python
# Crear universo cliente-producto-semana
# Fechas m√≠n/m√°x para generar semanas completas
min_week, max_week = df_transactions["week"].min(), df_transactions["week"].max()
weeks = pd.Series(df_transactions["week"].unique(), name="week")

# Transformar semanas a int16 para ahorrar memoria
weeks = weeks.astype("int16")

# Todos los clientes √ó productos √ó semanas
universe = (
    pd.MultiIndex.from_product(
        [df_customers["customer_id"].cat.categories,
         df_products["product_id"].cat.categories,
         weeks],
        names=["customer_id", "product_id", "week"]
    )
    .to_frame(index=False)
)

# Agregar las variables de clientes y productos
universe = universe.merge(df_customers, on="customer_id", how="left")
universe = universe.merge(df_products, on="product_id", how="left")
```

Ahora por √∫ltimo, eliminamos order_id, pues al ser simplemente un id no brinda informaci√≥n real al modelo, sin embargo, vemos que un cliente puede comprar el mismo producto varias veces la misma semana, por ende si aplicamos un drop duplicates sobre el conjunto de datos se perder√≠an esas instancias de compra, siendo que m√°s instancias de compra implican una mayor probabilidad de que por lo menos alguna compra (1) sea realizada por el cliente. Siendo esto √∫ltimo la definici√≥n misma de nuestra variable objetivo bought (no nos importa si compra 10 veces o una, con tal de que compre o no).

A su vez, para evitar data leakage removemos la variable items, ya que nos dice directamente si el producto fu√© bought o no (si es mayor a 0 implica bought). Tambi√©n al agregar week, purchase date es irrelevante pues el objetivo del estudio es semana a semana, por ende tambien removemos purchase_date.

```python

# Merge para target binario
data = universe.merge(df_grouped, on=["customer_id", "product_id", "week"], how="left")
data["bought"] = data["bought"].fillna(0).astype("int8")

# Fillna para average_items
data.drop(columns=["order_id", "purchase_date", "items"], inplace = True)

# Al dropear order_id, se repetir√≠an filas sise compra lo mismo varias veces a 
# la semana por un cliente
# Sin embargo, no eliminamos duplicados para darle el peso correspondiente 
# a varias compras semanales del mismo productos
# data = data.drop_duplicates()

data.head()
```

```python
# Ideas de nuevas features:
#   recency ‚Üí semanas desde la √∫ltima compra.
#   frequency ‚Üí n√∫mero de veces que compr√≥ en las √∫ltimas 6 semanas.
#   customer_product_share ‚Üí % de veces que este producto representa dentro de las compras del cliente.
#   trend ‚Üí comparaci√≥n entre compras recientes y antiguas.
def create_advanced_features(X: pd.DataFrame, y: pd.Series = None) -> pd.DataFrame:
    X_new = X.copy()
    if y is not None:
        X_new["bought"] = y.values

    X_new = X_new.sort_values(by=["customer_id", "product_id", "week"])

    # Recency
    X_new["last_purchase_week"] = X_new.groupby(["customer_id", "product_id"])["week"].shift(1)
    X_new["recency"] = X_new["week"] - X_new["last_purchase_week"]
    X_new.fillna({"recency": X_new["recency"].max()}, inplace=True)

    if "bought" in X_new.columns:
        # Frequency: √∫ltimas 6 semanas sin incluir la actual
        X_new["frequency"] = (
            X_new.groupby(["customer_id", "product_id"])["bought"]
            .transform(lambda s: s.shift(1).rolling(window=6, min_periods=1).sum())
        )

        # Total de compras del cliente hasta la semana anterior
        X_new["total_purchases"] = (
            X_new.groupby("customer_id")["bought"]
            .transform(lambda s: s.shift(1).cumsum())
        )

        # Customer-product share
        X_new["customer_product_share"] = (
            X_new["frequency"] / X_new["total_purchases"]
        )
        X_new["customer_product_share"].replace([np.inf, -np.inf], np.nan, inplace=True)
        X_new.fillna({"customer_product_share": 0}, inplace=True)

        # Trend (sin data leakage)
        recent = (
            X_new.groupby(["customer_id", "product_id"])["bought"]
            .transform(lambda s: s.shift(1).rolling(window=3, min_periods=1).sum())
        )
        past = X_new["frequency"] - recent
        X_new["trend"] = recent - past
    else:
        X_new["frequency"] = 0
        X_new["total_purchases"] = 0
        X_new["customer_product_share"] = 0
        X_new["trend"] = 0

    X_new.drop(columns=["bought", "last_purchase_week", "total_purchases"], errors="ignore", inplace=True)
    return X_new



# Aplicar feature engineering avanzado
X_train_feat = create_advanced_features(X_train_, y_train)
X_val_feat = create_advanced_features(X_val_, y_val)
X_test_feat = create_advanced_features(X_test_, y_test)
```

```python
# Manejo outliers
# Detecci√≥n outliers en variables num√©ricas (IQR)
def iqr_outlier_detection(X: pd.DataFrame) -> None:
    # Selecci√≥n de columnas num√©ricas
    X_num = X.select_dtypes(include=['float32', 'float64', 'int16', 'int32', 'int64'])
    continuous_cols = [col for col in X_num.columns if X_num[col].nunique() > 2]

    X_cont = X_num[continuous_cols]
    Q1 = np.percentile(X_cont, 25, axis=0)
    Q3 = np.percentile(X_cont, 75, axis=0)
    IQR = Q3 - Q1
    outliers = (X_cont < (Q1 - 1.5 * IQR)) | (X_cont > (Q3 + 1.5 * IQR))
    # Boxplot para cada variable num√©rica
    for i, col in enumerate(continuous_cols):
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=X_cont[col])
        plt.title(f"Boxplot de {col}")
        plt.xlabel(col)
        plt.show()
    
    print("N√∫mero de outliers por columna:")
    print(outliers.sum())
```

Esta estrategia permiti√≥ transformar la informaci√≥n at√≠pica en una feature significativa para el modelo, preservando patrones espaciales relevantes que podr√≠an estar relacionados con el comportamiento de compra de¬†los¬†clientes.

### PipeLine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, FunctionTransformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn import set_config

set_config(transform_output="pandas")

# =====================================================
# 1. Feature Engineering
# =====================================================

class FeatureEngineer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.y_train_ = None

    def fit(self, X, y=None):
        # guardamos y para usar en transform
        self.y_train_ = y.copy() if y is not None else None
        return self

    def transform(self, X, y=None):
        X_new = X.copy()
        y_to_use = y if y is not None else self.y_train_

        if y_to_use is not None:
            X_new["bought"] = y_to_use
            X_new = X_new.sort_values(by=["customer_id", "product_id", "week"])

            X_new["last_purchase_week"] = X_new.groupby(["customer_id", "product_id"])["week"].shift(1)
            X_new["recency"] = X_new["week"] - X_new["last_purchase_week"]
            X_new.fillna({"recency": X_new["recency"].max()}, inplace=True)

            X_new["frequency"] = (
                X_new.groupby(["customer_id", "product_id"])["bought"]
                .transform(lambda s: s.shift(1).rolling(window=6, min_periods=1).sum())
            )

            X_new["total_purchases"] = (
                X_new.groupby("customer_id")["bought"]
                .transform(lambda s: s.shift(1).cumsum())
            )

            X_new["customer_product_share"] = (
                X_new["frequency"] / X_new["total_purchases"]
            ).replace([np.inf, -np.inf], np.nan).fillna(0)

            recent = (
                X_new.groupby(["customer_id", "product_id"])["bought"]
                .transform(lambda s: s.shift(1).rolling(window=3, min_periods=1).sum())
            )
            past = X_new["frequency"] - recent
            X_new["trend"] = recent - past

            X_new.drop(columns=["bought", "last_purchase_week", "total_purchases", "customer_id"], inplace=True, errors="ignore")

        return X_new

    


# =====================================================
# 2. Clustering geogr√°fico
# =====================================================
class GeoClusterer(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=2):
        self.n_clusters = n_clusters
        self.kmeans = KMeans(n_clusters=self.n_clusters, random_state=1892)

    def fit(self, X, y = None):
        if not {"X", "Y"}.issubset(X.columns):
            raise ValueError("Se requieren las columnas 'X' y 'Y' para el clustering.")
        self.kmeans.fit(X[["X", "Y"]])
        return self

    def transform(self, X, y = None):
        if not {"X", "Y"}.issubset(X.columns):
            raise ValueError("Se requieren las columnas 'X' y 'Y' para el clustering.")
        X = X.copy()
        X["cluster"] = self.kmeans.predict(X[["X", "Y"]])
        return X

# =====================================================
# 3. Definici√≥n de transformaciones
# =====================================================
numerical_features = ["size", "num_deliver_per_week", "recency", "frequency", "customer_product_share", "trend"]
categorical_features = ["customer_type", "brand", "category", "sub_category", "segment", "package", "cluster", "product_id"]

numerical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("scaler", RobustScaler())
])

categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numerical_pipeline, numerical_features),
        ("cat", categorical_pipeline, categorical_features)
    ],
    remainder="passthrough",  # Mantener columnas no especificadas
    verbose_feature_names_out=False  # mantiene nombres legibles
)


# =====================================================
# 4. Pipeline final (Feature Engineering + Preprocesamiento)
# =====================================================
feature_engineering_pipeline = Pipeline([
    ("geo_clustering", GeoClusterer(n_clusters=2)),
    ("feature_engineering", FeatureEngineer()),
    ("preprocessor", preprocessor)
])

```

### Elecci√≥n de modelos a comparar y explicar sus hiperpar√°metros
#### Logistic Regression: 
* Tipo de modelo lineal-clasificaci√≥n probabil√≠stica. Modela la probabilidad de pertenecer a una clase como una combinaci√≥n lineal de las variables de entrada. Es r√°pido, interpretable y eficiente para relaciones aproximadamente lineales entre las features y la probabilidad de compra. 
* Principales hiperpar√°metros: 
    - C: inverso de la regularizaci√≥n. Valores m√°s altos reducen la regularizaci√≥n y permiten ajustar m√°s al entrenamiento.
    - penalty: tipo de regularizaci√≥n ('l1', 'l2', 'elasticnet', 'none').
    - class_weight: pondera las clases para manejar desbalance de datos.
    - max_iter: n√∫mero m√°ximo de iteraciones para convergencia.
* Ventajas: Simple, interpretable, r√°pido y estable.
* Desventajas: No captura relaciones no lineales ni interacciones complejas.

#### Decision Tree Classifier
* Tipo de modelo: √Årbol de decisi√≥n. Construye un √°rbol jer√°rquico de decisiones donde cada nodo divide los datos seg√∫n la feature que maximiza la separaci√≥n de clases (usualmente mediante Gini o Entropy). Cada rama representa una condici√≥n sobre las variables, y cada hoja predice la clase final.
* Principales hiperpar√°metros:
    - max_depth: profundidad m√°xima del √°rbol; controla la complejidad y riesgo de sobreajuste.
    - min_samples_split: n√∫mero m√≠nimo de muestras necesarias para dividir un nodo; previene divisiones poco significativas.
    - min_samples_leaf: n√∫mero m√≠nimo de muestras requeridas en una hoja; ayuda a suavizar el √°rbol.
    - max_features: n√∫mero m√°ximo de variables consideradas en cada divisi√≥n; puede reducir sobreajuste.
    - criterion: funci√≥n de calidad de la divisi√≥n ('gini' o 'entropy').
    - class_weight: ponderaci√≥n para clases desbalanceadas.
* Ventajas: F√°cil de interpretar y visualizar, y captura relaciones no lineales e interacciones entre variables.
* Desventajas: Alta varianza, propenso a sobreajuste si no se regula y sensible a peque√±os cambios en los datos.

#### Random Forest Classifier:
* Tipo de modelo de ensamble de √°rboles (Bagging). Crea m√∫ltiples √°rboles de decisi√≥n entrenados sobre muestras aleatorias de datos y variables; combina sus predicciones (votaci√≥n mayoritaria). Reduce la varianza del modelo individual y mejora la generalizaci√≥n.
* Principales hiperpar√°metros:
    - n_estimators: n√∫mero de √°rboles en el bosque.
    - max_depth: profundidad m√°xima de cada √°rbol (controla complejidad).
    - min_samples_split: n√∫mero m√≠nimo de muestras necesarias para dividir un nodo.
    - class_weight: ponderaci√≥n de clases para datos desbalanceados.
* Ventajas: Maneja relaciones no lineales, robusto al ruido y outliers.
* Desventajas: Menor interpretabilidad, puede ser m√°s lento con muchos √°rboles.

#### XGBClassifier (Extreme Gradient Boosting): 
* Tipo de modelo de ensamble (Boosting). Entrena √°rboles de manera secuencial, donde cada nuevo √°rbol corrige los errores del anterior. Optimiza una funci√≥n de p√©rdida mediante gradiente descendente, permitiendo capturar patrones complejos y no lineales.
* Principales hiperpar√°metros:
    - n_estimators: n√∫mero de √°rboles a entrenar.
    - learning_rate: controla cu√°nto contribuye cada √°rbol nuevo (menores valores requieren m√°s √°rboles).
    - max_depth: profundidad m√°xima de cada √°rbol.
    - subsample: fracci√≥n de observaciones usadas en cada iteraci√≥n (reduce sobreajuste).
    - colsample_bytree: fracci√≥n de variables consideradas por √°rbol.
* Ventajas: Alta precisi√≥n, captura relaciones complejas, muy usado en competencia.
* Desventajas: Mayor riesgo de sobreajuste si no se regula; requiere tuning fino.

```python
# Under-sampling para balancear clases
from sklearn.utils import resample
def undersample_data(X: pd.DataFrame, y: pd.Series, pond: int) -> (pd.DataFrame, pd.Series):
    # Combinar X e y para facilitar el resampling
    data = X.copy()
    data['bought'] = y

    # Separar clases mayoritaria y minoritaria
    majority = data[data['bought'] == 0]
    minority = data[data['bought'] == 1]

    # Subsample de la clase mayoritaria
    majority_downsampled = resample(majority,
                                    replace=False,
                                    n_samples=int(len(minority)*pond),
                                    random_state=1892)

    # Combinar de nuevo
    balanced_data = pd.concat([majority_downsampled, minority])

    # Mezclar los datos
    balanced_data = balanced_data.sample(frac=1, random_state=1892).reset_index(drop=True)

    # Separar X e y
    y_balanced = balanced_data['bought']
    X_balanced = balanced_data.drop(columns=['bought'])

    return X_balanced, y_balanced

X_train_balanced, y_train_balanced = undersample_data(X_train, y_train, pond=4)
```

```python
import optuna
import math
from optuna.samplers import TPESampler
from sklearn.metrics import recall_score, classification_report
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

optuna.logging.set_verbosity(optuna.logging.WARNING)

def objective_dt(trial):
    # Par√°metro a optimizar: multiplicador de under-sampling (negativos:positivos)
    undersample_multiplier = trial.suggest_int("undersample_multiplier", 1, 4)

    # Capeo para evitar pedir m√°s muestras que la clase mayoritaria
    maj = int((y_train == 0).sum())
    mino = int((y_train == 1).sum())
    max_pond_allowed = max(1, maj // max(1, mino))
    pond_eff = min(undersample_multiplier, max_pond_allowed)

    # Crear split balanceado seg√∫n el trial
    X_train_balanced, y_train_balanced = undersample_data(X_train, y_train, pond=pond_eff)

    # Preprocesamiento
    n_clusters = trial.suggest_int("geo_n_clusters", 2, 6)

    num_imputer_strategy = trial.suggest_categorical(
        "num_imputer_strategy", ["mean", "median", "most_frequent"]
    )

    cat_imputer_strategy = trial.suggest_categorical(
        "cat_imputer_strategy", ["most_frequent", "constant"]
    )
    fill_value = "missing" if cat_imputer_strategy == "constant" else None

    scaler_choice = trial.suggest_categorical("scaler", ["robust", "standard", "none"])
    if scaler_choice == "robust":
        scaler = RobustScaler()
    elif scaler_choice == "standard":
        scaler = StandardScaler()
    else:
        scaler = "passthrough"

    min_freq = trial.suggest_float("min_frequency", 0.0, 0.05)

    numerical_pipeline_trial = Pipeline([
        ("imputer", SimpleImputer(strategy=num_imputer_strategy)),
        ("scaler", scaler)
    ])

    categorical_pipeline_trial = Pipeline([
        ("imputer", SimpleImputer(strategy=cat_imputer_strategy, fill_value=fill_value)),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False, min_frequency=min_freq))
    ])

    preprocessor_trial = ColumnTransformer(
        transformers=[
            ("num", numerical_pipeline_trial, numerical_features),
            ("cat", categorical_pipeline_trial, categorical_features)
        ],
        remainder="passthrough",
        verbose_feature_names_out=False
    )

    # Hiperpar√°metros del √°rbol
    dt_params = {
        "criterion": trial.suggest_categorical("dt_criterion", ["gini", "entropy"]),
        "splitter": trial.suggest_categorical("dt_splitter", ["best", "random"]),
        "max_depth": trial.suggest_int("dt_max_depth", 2, 40),
        "min_samples_split": trial.suggest_int("dt_min_samples_split", 2, 200),
        "min_samples_leaf": trial.suggest_int("dt_min_samples_leaf", 1, 100),
        "max_features": trial.suggest_categorical("dt_max_features", [None, "sqrt", "log2", 0.5, 0.8, 1.0]),
        "class_weight": trial.suggest_categorical("dt_class_weight", [None, "balanced"]),
        "random_state": 1892,
    }

    model = DecisionTreeClassifier(**dt_params)

    pipeline = Pipeline([
        ("geo_clustering", GeoClusterer(n_clusters=n_clusters)),
        ("feature_engineering", FeatureEngineer()),
        ("preprocessor", preprocessor_trial),
        ("classifier", model)
    ])

    pipeline.fit(X_train_balanced, y_train_balanced)
    preds = pipeline.predict(X_val)

    score = recall_score(y_val, preds)
    trial.set_user_attr("pipeline", pipeline)
    trial.set_user_attr("undersample_multiplier_eff", pond_eff)
    return score
```
